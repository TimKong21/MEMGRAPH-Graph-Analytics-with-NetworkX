## Import necessary module

```python
import networkx as nx
import matplotlib.pyplot as plt
import pprint as pp
```

## Dijkstraâ€™s algorithm from node 0 to node 4

```python
g = nx.read_edgelist('graph-assignment.txt', nodetype=str,
                     data=[('weight', int)], create_using=nx.Graph())

pos = {'0':[-1,0], '1':[-0.5,0.5], '8':[-0.5,-0.5],
       '2':[0,0.5], '6':[0,0], '7':[0,-0.5],
       '3':[0.5,0.5], '4':[1,0], '5':[0.5,-0.5]}

# draw hex-like graph with fixed positions
nx.draw(g, pos, with_labels=True, node_color="#FF4133")

# draw edge label
edge_labels = nx.get_edge_attributes(g, 'weight')
nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels)

p0to4 = nx.shortest_path(g, source='0', target='4', weight='weight')
p0to4_length = nx.shortest_path_length(g, source='0', target='4', weight='weight')
print("The shortest path from node 0 to 4:\n", p0to4, "With length of:", p0to4_length)

ax= plt.gca()
ax.collections[0].set_edgecolor("#000000") #set the black node border
plt.show()
```

## Construct a tree with the BFS algorithm starting from node 7

```python
g = nx.read_edgelist('graph-assignment.txt', nodetype=str,
                     data=[('weight', int)], create_using=nx.Graph())

pos = nx.planar_layout(g)

# draw graph with planar layout
nx.draw(g, pos, with_labels=True, node_color="#FF4133")

# draw tree 
bfs = nx.bfs_tree(g, source="7")
nx.draw(bfs, pos, with_labels=True, node_color="#f86e00", edge_color="#5B33FF")

plt.show()
```

## Use the Page Rank algorithm to find the node with the highest ranking

```python
g = nx.read_edgelist('graph-assignment.txt', nodetype=str,
                     data=[('weight', int)], create_using=nx.Graph())

pos = nx.spring_layout(g)

# page rank the ndes
pr = nx.pagerank(g)
pp.pprint(pr)

# get node with max page rank
key_list = list(pr.keys())
val_list = list(pr.values())
max_pr = max(val_list)
position = val_list.index(max_pr)
max_pr_node = key_list[position]

# make node with hiughest rank yelllow,and red for the others
node_color = []
for node in g.nodes:
    if node == max_pr_node:
        node_color.append('#F9FF33')
    else:
        node_color.append('#FF4133')

# draw graph with planar layout
nx.draw(g, with_labels=True, node_color=node_color)

ax= plt.gca()
ax.collections[0].set_edgecolor("#000000") #set the black node border
plt.show()
```
