## Import necessary modules

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Read `.grahml` file and explore attributes

```python
G = nx.read_graphml('graph.graphml.txt')

print("node attributes")
for node_id in G.nodes:
    print("node {0}: {1}".format(node_id, G.nodes[node_id]))
    
print("\nedge attributes")
for edge_id in G.edges:
    print("edge {0}: {1}".format(edge_id, G.edges[edge_id])) 

print("\ngraph attributes")
G.graph
```

## Customize node attributes

```python
node_color = []
node_size = []
#node attributes
for node_id in G.nodes:
    label = G.nodes[node_id]['label']
    if label == 'Person':
        node_color.append('red')
        node_size.append(5000)
        
    elif label == 'Store':
        node_color.append('orange')
        node_size.append(3000)
        
    elif label == 'Pos':
        node_color.append("white")
        node_size.append(2000)
        
    else:
        node_color.append("white")
        node_size.append(10000)
```

## Customioze edge attributes

```python
edge_color = []
width=[]
#edge attrtibutes
for edge_id in G.edges:
    if edge_id == ('1', '2'):
        edge_color.append('purple')
        width.append(4)
        
    else:
        edge_color.append('black')
        width.append(1)
```

## Customize nodes positions

```python
positions = {'1':[-1,0.5], '2':[-0.5,0.6], '22':[-0.5,0.4],
            '6':[0,0.6], '3':[0,0.45], '5':[0.5,0.55], '4':[0.5,0.45]}
```

## Visualize graph

```python
plt.figure(figsize=(8,5)) 

nx.draw_networkx(G,
                 positions,
                 labels=nx.get_node_attributes(G, 'label'),  
                 font_size=13,
                 node_size=node_size,
                 node_color=node_color,
                 edge_color=edge_color,
                 arrowsize =20,
                 width=width)
ax= plt.gca()
ax.collections[0].set_edgecolor("#000000") #set the black node border
plt.savefig('Assignment 2.png')
plt.show()
```
