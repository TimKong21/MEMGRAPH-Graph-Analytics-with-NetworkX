## Import necessary modules

```python
import networkx as nx
import matplotlib.pyplot as plt
import pprint as pp
```

## Graph traversal

### Breadth-First Search 

```python
# read all lines from graph.txt
with open('graph.txt') as f:
    lines = f.readlines()

# strip (remove leading & tailing white space)
# and split(remove with white space)
# list comprehension
edgeList = [line.strip().split() for line in lines] 

g = nx.Graph()
g.add_edges_from(edgeList)
print(g.nodes)

pos = nx.planar_layout(g)
nx.draw(g, pos, with_labels=True, node_color="#f86e00")

bfs = nx.bfs_tree(g, source="go")
nx.draw(bfs, pos, with_labels=True, node_color="#f86e00", edge_color="#dd2222")

plt.show()
```

### Depth-First Search

```python
with open('graph.txt') as f:
    lines = f.readlines()

edgeList = [line.strip().split() for line in lines]

g = nx.Graph()
g.add_edges_from(edgeList)
print(g.nodes)

pos = nx.planar_layout(g)
nx.draw(g, pos, with_labels=True, node_color="#f86e00")

dfs = nx.dfs_tree(g, source="go")
nx.draw(dfs, pos, with_labels=True, node_color="#f86e00", edge_color="#dd2222")

plt.show()
```

## Path Finding algorithim 

### Dijkstra

```python
# file has weight embeded - specify data -> column 3 is all int + name it as 'weight'
# nx knows how to read it
g = nx.read_edgelist('graph-weighted.txt', nodetype=str,
                     data=[('weight', int)], create_using=nx.Graph())

pos = nx.planar_layout(g)
nx.draw(g, pos, with_labels=True, node_color="#f86e00")

edge_labels = nx.get_edge_attributes(g, 'weight')
nx.draw_networkx_edge_labels(g, pos,edge_labels=edge_labels)

p1to6 = nx.shortest_path(g, source="go", target="i", weight='weight') #shortest route between "go" and "i"
p1 = nx.shortest_path(g, source="go", weight='weight') #shortest route between "go" and each nodes

length = nx.shortest_path_length(g, source="go", target="i", weight='weight') #shortest distance between "go" and "i"
lengths = nx.shortest_path_length(g, source="go", weight='weight') #shortest distance between "go" and each nodes

print("The shortest path from GO to each node: ")
results = {a: list(b) for a, b in p1.items()}
pp.pprint(results) #prettifier
print("\nThe shortest path length from Go to each node:\n",lengths)
print("\nThe shortest path from GO to i:\n", p1to6, "With length of:", length)

plt.show()
```

## Centralities

```python
def draw_network(G, layout, colors, position, draw_labels, title):

    ax = plt.subplot(position)
    ax.title.set_text(title)

    nx.draw(G,
            pos=layout,
            node_size=50,
            node_color=colors,
            edge_color='g')

    if draw_labels:
        node_labels = nx.get_node_attributes(G, 'label')
        nx.draw_networkx_labels(G, pos=layout, labels=node_labels)

G = nx.read_gexf('../../Graph Analytics with NetworkX/Lecture 2_Other visualization tools/graph.gexf.txt')

# Degree centrality
centrality = nx.degree_centrality(G)
colors = list(centrality.values())
draw_network(G, nx.spring_layout(G), colors, 221,
             draw_labels=False, title='Degree centrality')

# Closeness centrality
centrality = nx.closeness_centrality(G)
colors = list(centrality.values())
draw_network(G, nx.spring_layout(G), colors, 222,
             draw_labels=False,  title='Closeness centrality')

# Betweness centrality
centrality = nx.betweenness_centrality(G)
colors = list(centrality.values())
draw_network(G, nx.spring_layout(G), colors, 223,
             draw_labels=False,  title='Betweenness centrality')

# Katz centrality
centrality = nx.katz_centrality(G)
colors = list(centrality.values())
draw_network(G, nx.spring_layout(G), colors, 224,
             draw_labels=False,  title='Katz centrality')

plt.show()
```

## Link Analysis and Page Rank

### Page Rank 

```python
with open('graph-pagerank.txt') as f:
    lines = f.readlines()

edgeList = [line.strip().split() for line in lines]

G = nx.DiGraph()
G.add_edges_from(edgeList)

ppr1 = nx.pagerank(G)

pp.pprint(ppr1)

pos = nx.planar_layout(G)
nx.draw(G, pos, with_labels=True, node_color="#f86e00")
plt.show()
```

```python
G = nx.read_gexf('../../Graph Analytics with NetworkX/Lecture 2_Other visualization tools/graph.gexf.txt')

pagerank = nx.pagerank(G)

# add new attributes
for node_id in G.nodes:
    G.nodes[node_id]['pagerank'] = round(pagerank[node_id], 5)

# aprint existing attributes
for node_id in G.nodes:
    print("node {0}: {1}".format(node_id, G.nodes[node_id]))

pos = nx.spring_layout(G)
nx.draw(G, pos)

node_labels = nx.get_node_attributes(G, 'pagerank')
nx.draw_networkx_labels(G, pos, labels=node_labels)

plt.show()
```
